# GroupBy con PySpark

Funciona igual que con pandas

```python
df = spark.read.csv('/FileStore/tables/datos.csv',sep = ';', header = True, inferSchema =True)
```

```python
df = df.na.drop(how='any')
```

```python
df.show(25)
```

Vamos a ver cuantos pacientes han ido a cada hospital

```python
df.groupBy('Hospital').count('Name').show()
```

Mostremos el total de horas y de facturación por hospital

```python
df.groupBy('Hospital').sum('Tiempo_en_hospital_horas','Facturacion').show()
```

Mostremos la media de edad de los pacientes de cada hospital

```python
df.groupBy('Hospital').avg('Edad').show()
```

Por otro lado, para guardar el Tabla con los datos depurados usaremos el siguiente comando

```python
df.write.csv('datos_final.csv',header=True)
```

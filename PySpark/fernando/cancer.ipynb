```
# Importamos las librerías necesarias
from pyspark.sql.functions import col, avg, round
```

```
#  /FileStore/tables/datos.csv

# Leemos el CSV

df = spark.read.csv('/FileStore/tables/datos.csv',sep = ';', header = True, inferSchema =True)
```

```
# Echamos un primer vistazo a los datos
df.show()
```

```
df.printSchema()
```

```
df.show()
```

```
# Borramos los valores nulos
df= df.na.drop(how='any')
```

```
# Observamos los valores que quedan
df.count()
```

```
# Creamos la columna Dias_en_el_hospital en una nueva variable y redondeamos el resultado
df2 = df.withColumn('Dias_en_el_hospital', round(df['Tiempo_en_hospital_horas']/24, 2))
```

```
#Observamos los cambios
df2.show()
```

```
# Creamos la columna Facturacion_por_hora en una nueva variable y redondeamos el resultado
df3 = df2.withColumn('Facturacion_por_hora', round(df2['Facturacion']/df2['Tiempo_en_hospital_horas'], 2))
```

```
# Observamos los cambios
df3.show()
```

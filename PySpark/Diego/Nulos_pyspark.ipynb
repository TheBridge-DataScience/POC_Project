```
df=spark.read.csv('dbfs:/FileStore/tables/datos.csv', header=True, inferSchema=True, sep=';')
```

```
df.show(25)
```

```
from pyspark.sql.functions import col   #Para identificar las filas con nulos.

filas_nulas = df.filter(col(df.columns[0]).isNull())
for columna in df.columns[1:]:
    filas_nulas = filas_nulas.union(df.filter(col(columna).isNull()))

filas_nulas = filas_nulas.distinct()

filas_nulas.show()
```

```
df.na.drop(how = 'any', thresh = 2).show(25)  #Aquí estamos dejando las filas que al menos tengan 2 valores No nulos.
```
